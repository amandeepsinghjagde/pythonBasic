```python
import seaborn as sns
```

```python
tips= sns.load_dataset("tips")
```

```python
tips.head()
```

```python
tips.info()
```

```python
tips.isna().sum() # checking NA values
```

```python
sns.distplot(tips["total_bill"])
```

```python
sns.distplot(tips["size"], kde=False, color="red") #kernal density estimator
```

```python
sns.jointplot("total_bill","tip",data = tips, color = "m") #scatter plot
```

```python
sns.jointplot("total_bill","tip",data = tips, color = "m", kind = "reg") #regeration 
```

```python
sns.jointplot("total_bill","tip",data = tips, color = "g", kind = "hex") #hexagon
```

```python
sns.barplot(x="sex",y="total_bill",data=tips) #estimator is always mean by default
```

```python
sns.countplot(x="sex",data=tips) #checking the number of male and female
```

```python
sns.countplot(x="sex",data=tips, hue="day",palette="rainbow") #its to make it more informative , hue should be a catagorical 
```

```python
sns.boxplot(x="day",y="total_bill",data=tips, hue ="sex", palette= "rainbow")
```

```python
tc= tips.corr()
```

```python
tc
```

```python
sns.heatmap(data=tc)
```

```python
sns.heatmap(data=tc, annot=True , cmap="plasma", linecolor="white",linewidths=5) #heatmap with some paramters
```

```python
%matplotlib inline  #used to view plots on jupitor notebook not need to write plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(10)) # to draw simple line
```

```python
year = [1950,1970,1990,2000]
pop = [2.13,3.07,3.8,4.09]
```

```python
plt.scatter(year,pop)
```

```python
plt.title("World Population")
plt.xlabel("Years")
plt.ylabel("Pop")
plt.plot(year,pop)
```

```python
plt.hist(pop) # histogram
```

```python
from matplotlib import style
```

```python
plt.style.use("grayscale")
plt.title("World Population")
plt.xlabel("Years")
plt.ylabel("Pop")
plt.plot(year,pop)
```

```python
print(plt.style.available)
```

```python
plt.style.use("dark_background")
plt.title("World Population")
plt.xlabel("Years")
plt.ylabel("Pop")
plt.plot(year,pop,color="green",linestyle=":",label="line1", linewidth=4)
plt.legend()
```

```python
values = [12,55,4,32,15]
colours = ["red","blue","green","cyan","m"]
explode = [0,0,0.5,0.4,0]
labels = ["India","US","China","Russia","Europe"]
plt.pie(values,colors=colours,labels=labels, autopct="%1.3f%%", explode= explode,)
plt.title("Student Locations")
plt.savefig("Student_country.png")
```
